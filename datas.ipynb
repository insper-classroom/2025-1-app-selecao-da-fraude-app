```python
import pandas as pd
import numpy as np
```

```python
df_pays = pd.read_feather("data/payers-v1.feather")
df_pays
```

```python
df_sellers = pd.read_feather("data/seller_terminals-v1.feather")
df_sellers
```

```python
df_transactions = pd.read_feather("data/transactions_train-v1.feather")
df_transactions
```

```python
# Verificando as colunas de cada dataframe para identificar como realizar o merge
print("Colunas em df_pays:")
print(df_pays.columns.tolist())
print("\nColunas em df_sellers:")
print(df_sellers.columns.tolist())
print("\nColunas em df_transactions:")
print(df_transactions.columns.tolist())
```

```python
# Verificando as primeiras linhas de cada dataframe para entender a estrutura dos dados
print("Primeiras linhas de df_pays:")
display(df_pays.head(2))
print("\nPrimeiras linhas de df_sellers:")
display(df_sellers.head(2))
print("\nPrimeiras linhas de df_transactions:")
display(df_transactions.head(2))
```

```python
# Realizando o merge dos trÃªs dataframes
# Com base na estrutura dos dados, identificamos que:
# - df_pays tem 'card_hash' que parece corresponder a 'card_id' em df_transactions
# - df_transactions tem 'terminal_id' que se relaciona com 'terminal_id' em df_sellers

df_merged = pd.merge(df_transactions, df_pays, left_on='card_id', right_on='card_hash', how='left')
df_merged = pd.merge(df_merged, df_sellers, on='terminal_id', how='left')

print(f"DimensÃµes do dataframe final: {df_merged.shape}")
print(f"NÃºmero original de transaÃ§Ãµes: {len(df_transactions)}")
df_merged.head()
```

```python
df_merged.to_csv("data/merged_dataset.csv", index=False)
print("Dataset combinado tambÃ©m salvo como 'merged_dataset.csv'")
```
